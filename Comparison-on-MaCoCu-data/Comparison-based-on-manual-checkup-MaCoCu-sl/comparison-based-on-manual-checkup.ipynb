When I performed manual checkup of the most common domains in the MaCoCu-sl corpus, I also annotated them with domain genres.
I will check whether these domains are present in our sample as well and how they are connected with classifiers predictions.

```python
import pandas as pd
```

```python
# Import the list with annotated domains
manual_an_df = pd.read_csv("genres-in-Slovene-domains.csv")

manual_an_df.head(2)
```

```python
# Rename the column web domain
manual_an_df = manual_an_df.rename(columns={"domain": "domain_id"})
manual_an_df.head(1)
```

```python
manual_an_df.shape
```

```python
# Open the main file with predictions on domain level
df = pd.read_csv("MaCoCu-sample2-with-predictions-genre-distribution-in-domains.csv", sep="\t", index_col = 0)
df.head(2)
```

```python
df.describe(include="all")
```

We have 300 manually annotated domains and 1000 domains in the sample. We'll see how many of these 300 appear in the sample.

```python
# Merge the two dfs
df_merged = pd.merge(left= df, right = manual_an_df, on = "domain_id", how = "inner" )

df_merged
```

```python
df_merged[['domain_id', 'Genre', 'most_frequent_label_FTD',
       'most_frequent_label_GINCO', 'most_frequent_label_CORE', 
       'most_frequent_label_X-GENRE' ]]
```
