## Apply mapping to GINCO sheet

````python
# Apply mapping to the sheet with all information

GINCO_sheet = pd.read_csv("data-sheets-with-all-info/GINCO-MT-GINCO-keeptext-with-all-information.csv", sep = "\t", index_col= 0)
GINCO_sheet.head(5)
````

````python
GINCO_sheet["primary_level_1_to_X-GENRE"] = GINCO_sheet["primary_level_1"].map(map_GINCO)

GINCO_sheet["primary_level_1_to_X-GENRE"]
````

````python
# Check whether there is any NaN value in the column with mappings GINCO_sheet
GINCO_sheet[GINCO_sheet["primary_level_1_to_X-GENRE"].isnull().values]
````

````python
# Save the new GINCO sheet
GINCO_sheet.to_csv("data-sheets-with-all-info/GINCO-MT-GINCO-keeptext-with-all-information.csv", sep = "\t")
````

````python
# Analyze distribution of labels according to the new mapping
count = pd.DataFrame({"Count": list(GINCO_sheet["primary_level_1_to_X-GENRE"].value_counts()), "Percentage": list(GINCO_sheet["primary_level_1_to_X-GENRE"].value_counts(normalize=True))}, index = GINCO_sheet["primary_level_1_to_X-GENRE"].value_counts().index)

print(count.to_markdown())
````

## Apply mapping to FTD sheet

````python
# Apply mapping to the sheet with all information

FTD_sheet = pd.read_csv("data-sheets-with-all-info/FTD-dataset-with-all-information.csv", sep = "\t", index_col= 0)
FTD_sheet.head(5)
````

````python
FTD_sheet.columns
````

````python
FTD_sheet["labels_to_X-GENRE"] = FTD_sheet["labels"].map(map_FTD).fillna("check")

FTD_sheet["labels_to_X-GENRE"]
````

````python
# Check whether there is any NaN value in the column with mappings
# - there are some Nan values, but they are assigned to texts with multiple labels which is ok
Nan_values = FTD_sheet[FTD_sheet["labels_to_X-GENRE"] == "check"]
Nan_values
Nan_values[Nan_values["Multiple labels"] != "y"]
````

````python
# Save the new sheet
FTD_sheet.to_csv("data-sheets-with-all-info/FTD-dataset-with-all-information.csv", sep = "\t")
````

````python
# Analyze distribution of labels according to the new mapping
count = pd.DataFrame({"Count": list(FTD_sheet["labels_to_X-GENRE"].value_counts()), "Percentage": list(FTD_sheet["labels_to_X-GENRE"].value_counts(normalize=True))}, index = FTD_sheet["labels_to_X-GENRE"].value_counts().index)

print(count.to_markdown())
````

````python
# Discard the "discarded" labels and those that have "check"
# (in the case of FTD, these are the texts with multiple labels)
# to calculate distribution without them
FTD_sheet = FTD_sheet[FTD_sheet["labels_to_X-GENRE"] != "discarded"]
FTD_sheet = FTD_sheet[FTD_sheet["labels_to_X-GENRE"] != "check"]

FTD_sheet.describe()
````

````python
count_FTD = pd.DataFrame({"Count": list(FTD_sheet["labels_to_X-GENRE"].value_counts()), "Percentage": list(round(FTD_sheet["labels_to_X-GENRE"].value_counts(normalize=True)*100, 2))}, index = FTD_sheet["labels_to_X-GENRE"].value_counts().index)

print(count_FTD.to_markdown())
````

## Apply mapping to CORE

````python
# Apply mapping to the sheet with all information

CORE_sheet = pd.read_csv("data-sheets-with-all-info/CORE-all-information.csv", sep = "\t", index_col= 0)
CORE_sheet.head(5)
````

````python
CORE_sheet.columns
````

````python
CORE_sheet["subcategories_to_X-GENRE"] = CORE_sheet["full_names"].map(map_CORE).fillna("check")

CORE_sheet["subcategories_to_X-GENRE"]
````

````python
# Check whether there is any NaN value in the column with mappings
# - there are some Nan values, but they are assigned to texts with multiple labels which is ok
Nan_values = CORE_sheet[CORE_sheet["subcategories_to_X-GENRE"] == "check"]

Nan_values["sub_len"].value_counts()
````

````python
# Save the new CORE sheet
CORE_sheet.to_csv("data-sheets-with-all-info/CORE-all-information.csv", sep = "\t")
````

````python
# Analyze distribution of labels according to the new mapping
count = pd.DataFrame({"Count": list(CORE_sheet["subcategories_to_X-GENRE"].value_counts()), "Percentage": list(CORE_sheet["subcategories_to_X-GENRE"].value_counts(normalize=True))}, index = CORE_sheet["subcategories_to_X-GENRE"].value_counts().index)

print(count.to_markdown())
````

````python
# Discard the "discarded" labels and those that have "check"
# (in the case of CORE, these are the texts with multiple labels or no CORE subcategory label)
# to calculate distribution without them
CORE_sheet = CORE_sheet[CORE_sheet["subcategories_to_X-GENRE"] != "discarded"]
CORE_sheet = CORE_sheet[CORE_sheet["subcategories_to_X-GENRE"] != "check"]

CORE_sheet.describe()
````

````python
# Analyze distribution of labels according to the new mapping
count_CORE = pd.DataFrame({"Count": list(CORE_sheet["subcategories_to_X-GENRE"].value_counts()), "Percentage": list(round(CORE_sheet["subcategories_to_X-GENRE"].value_counts(normalize=True)*100, 2))}, index = CORE_sheet["subcategories_to_X-GENRE"].value_counts().index)

print(count_CORE.to_markdown())
````
