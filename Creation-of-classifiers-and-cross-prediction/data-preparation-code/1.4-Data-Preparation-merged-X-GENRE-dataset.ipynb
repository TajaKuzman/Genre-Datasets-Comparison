```python
# Here, we will merge the train-dev-test splits from separate datasets into one joint train-dev-test split.
```

```python
import pandas as pd
```

```python
# Import datasets
core_train = pd.read_csv("C:/Users/tajak/Google Drive/GitHub/Genre-Datasets-Comparison/data-splits/CORE-sample-X-GENRE-train.csv", index_col = 0)
core_test = pd.read_csv("C:/Users/tajak/Google Drive/GitHub/Genre-Datasets-Comparison/data-splits/CORE-sample-X-GENRE-test.csv", index_col = 0)
core_dev = pd.read_csv("C:/Users/tajak/Google Drive/GitHub/Genre-Datasets-Comparison/data-splits/CORE-sample-X-GENRE-dev.csv", index_col = 0)

core_train
```

```python
ginco_train = pd.read_csv("C:/Users/tajak/Google Drive/GitHub/Genre-Datasets-Comparison/data-splits/GINCO-SL-X-GENRE-train.csv", index_col = 0)
ginco_test = pd.read_csv("C:/Users/tajak/Google Drive/GitHub/Genre-Datasets-Comparison/data-splits/GINCO-SL-X-GENRE-test.csv", index_col = 0)
ginco_dev = pd.read_csv("C:/Users/tajak/Google Drive/GitHub/Genre-Datasets-Comparison/data-splits/GINCO-SL-X-GENRE-dev.csv", index_col = 0)

ginco_train
```

```python
ftd_train = pd.read_csv("C:/Users/tajak/Google Drive/GitHub/Genre-Datasets-Comparison/data-splits/FTD-X-GENRE-train.txt", index_col = 0, sep="\t")
ftd_test = pd.read_csv("C:/Users/tajak/Google Drive/GitHub/Genre-Datasets-Comparison/data-splits/FTD-X-GENRE-test.txt", index_col = 0, sep="\t")
ftd_dev = pd.read_csv("C:/Users/tajak/Google Drive/GitHub/Genre-Datasets-Comparison/data-splits/FTD-X-GENRE-dev.txt", index_col = 0, sep="\t")

ftd_train
```

```python
ftd_dev
```

```python
# Merge splits
train1 = pd.concat([core_train, ginco_train])

train1
```

```python
final_train = pd.concat([train1, ftd_train])

final_train
```

```python
final_train.reset_index(drop= True, inplace = True)
final_train
```

```python
# Save the file
final_train.to_csv("C:/Users/tajak/Google Drive/GitHub/Genre-Datasets-Comparison/data-splits/X-GENRE-train.csv", sep="\t")
```

```python
# Merge splits
dev1 = pd.concat([core_dev, ginco_dev])

dev1
```

```python
final_dev = pd.concat([dev1, ftd_dev])

final_dev
```

```python
final_dev.reset_index(drop= True, inplace = True)
final_dev
```

```python
# Save the file
final_dev.to_csv("C:/Users/tajak/Google Drive/GitHub/Genre-Datasets-Comparison/data-splits/X-GENRE-dev.csv", sep="\t")
```

```python
# Merge splits
test1 = pd.concat([core_test, ginco_test])

test1
```

```python
final_test = pd.concat([test1, ftd_test])

final_test
```

```python
final_test.reset_index(drop= True, inplace = True)
final_test
```

```python
# Save the file
final_test.to_csv("C:/Users/tajak/Google Drive/GitHub/Genre-Datasets-Comparison/data-splits/X-GENRE-test.csv", sep="\t")
```

```python
# Merge all three splits to calculate label distribution
entire1 = pd.concat([final_train, final_dev])
entire1
```

```python
joined_splits = pd.concat([entire1, final_test])

joined_splits
```

```python
# Analyze distribution of labels according to the new mapping
count = pd.DataFrame({"Count": list(joined_splits["labels"].value_counts()), "Percentage": list(joined_splits["labels"].value_counts(normalize=True))}, index = joined_splits["labels"].value_counts().index)

print(count.to_markdown())
```
