## Compare CORE-sub labels with GINCO

```python
import pandas as pd
import numpy as np
```

```python
# Import the GINCO dataset
ginco = pd.read_csv("data-sheets-with-all-info/GINCO-MT-GINCO-keeptext-with-all-information.csv", sep="\t", index_col=0)
ginco.head(3)
```

```python
# Analyse how GINCO primary labels are connected with CORE labels based on the prediction
pd.crosstab(ginco["primary_level_1"], ginco["CORE_sub_pred_on_SL"], normalize="index")
```

```python
pd.crosstab(ginco["primary_level_1"], ginco["CORE_sub_pred_on_SL"], normalize="index").to_dict("index")
```

## Compare CORE-sub labels with FTD

```python
# Import the FTD dataset
ftd = pd.read_csv("data-sheets-with-all-info/FTD-dataset-with-all-information.csv", sep="\t", index_col=0)
ftd.head(3)
```

```python
ftd.columns
```

```python
ftd["Multiple labels"].value_counts()
```

```python
# Discard texts with multiple labels
ftd = ftd[ftd["Multiple labels"] != "y"]
ftd.describe(include="all")
```

```python
# Analyse how CORE main labels are connected with FTD labels based on the prediction
pd.crosstab(ftd['labels'], ftd["CORE_sub_pred"], normalize="index")
```

```python
pd.crosstab(ftd['labels'], ftd["CORE_sub_pred"], normalize="index").to_dict("index")
```
